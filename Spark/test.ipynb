```python
from tarfile import data_filter

import findspark
from pyspark.sql.utils import pyspark_column_op

findspark.init()

import pyspark

print(pyspark.__version__)
```

```python
# %% Configure Spark
# Set application name and master (local mode)
conf = (
        pyspark.SparkConf().setAppName("PySparkCSVReading").setMaster("local[2]")
)  # Using 2 cores locally
sc = pyspark.SparkContext(conf=conf)

# %% Load and process CSV data
try:
    file_path = "recent-grads.csv"  # Ensure the file path is correct
    rdd = sc.textFile(file_path)  # Load CSV file
    data = rdd.map(lambda line: line.split(","))  # Split lines into columns
    header = data.first()  # Extract header
    rows = data.filter(lambda x: x != header)  # Exclude header
    print(f"Header: {header}")
    print(f"First row: {rows.take(1)}")
except Exception as e:
    print(f"Error: {e}")
finally:
    sc.stop()  # St
```
